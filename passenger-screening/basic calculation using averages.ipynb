```python
# read in data
import pandas as pd
X = pd.read_csv('/home/jeff/Downloads/data/passenger-screening/stage1_labels.csv')
X_dict = {}
for row in X.itertuples():
    scan_id, zone = row.Id.split('_')
    zone_num = int(zone[4:])
    if scan_id not in X_dict:
        X_dict[scan_id] = [0]*17
    X_dict[scan_id][zone_num-1] = row.Probability
X = pd.DataFrame.from_dict(X_dict, 'index')
X.columns = ['Zone1', 'Zone2', 'Zone3', 'Zone4', 'Zone5', 'Zone6', 'Zone7', 'Zone8',
            'Zone9', 'Zone10', 'Zone11', 'Zone12', 'Zone13', 'Zone14', 'Zone15',
            'Zone16', 'Zone17']
X.head()
```

```python
# calculate mean of each column
means = X.mean()
means
```

```python
# determine ids of scans for which to make predictions
sample_submission = pd.read_csv('/home/jeff/Downloads/data/passenger-screening/stage1_sample_submission.csv')
submission_ids = []
for index in sample_submission.iloc[:,0]:
    scan_id = index.split('_')[0]
    if scan_id not in submission_ids:
        submission_ids.append(scan_id)
len(submission_ids)
```

```python
# use means to make predictions
with open('/home/jeff/Downloads/data/passenger-screening/my_predictions.csv', 'w') as pred_file:
    pred_file.write('Id,Probability\n')
    for scan_id in submission_ids:
        for zone_num in range(1,18):
            value = means[zone_num-1]
            pred_file.write('%s_Zone%d,%f\n' % (scan_id,zone_num,value))
```
